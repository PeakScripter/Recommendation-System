```python
ratings_name = big.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratings_name.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
ratings_name
```

```python
avg_rating = big.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating
```

```python
pop_df = ratings_name.merge(avg_rating, on='Book-Title')
pop_df
```

```python
# popular page

pop_df = pop_df[pop_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False)
pop_df
```

```python
x = big.groupby("User-ID").count()['Book-Rating']>200
imp_users = x[x].index
```

```python
filterrating = big[big['User-ID'].isin(imp_users)]
# filterrating
```

```python
y = filterrating.groupby("Book-Title").count()["Book-Rating"]>=50
famousbook = y[y].index
# famousbook
```

```python
final_rating = filterrating[filterrating["Book-Title"].isin(famousbook)]
# final_rating
```

```python
chart = final_rating.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
# chart
```

```python
chart.fillna(0,inplace=True)
chart
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
hehe = cosine_similarity(chart)
hehe
```

```python
hehe.shape
```

```python
def recommend(book_name):
    index = np.where(chart.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(hehe[index])),key=lambda x:x[1],reverse=True)[1:5]
    data=[]
    for i in similar_items:
        item=[]
        temp_df = books[books['Book-Title']==chart.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend("1984")
```
